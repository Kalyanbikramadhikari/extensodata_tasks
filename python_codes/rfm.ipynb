```python
#convert tran_date to default datetime
df['tran_date'] = pd.to_datetime(df['tran_date'], format='%d/%m/%Y', errors='coerce' )
df.head()
```

```python
#calculate recency


# Convert date_today to datetime
date_today = '05-02-2024'
date_today = pd.to_datetime(date_today)


recency = df.groupby(['account_number']).agg({"tran_date": lambda x: (date_today - x.max()).days })
recency.head()
```

```python
#frequency
frequency= df.groupby(['account_number']).agg({'account_number':lambda x:len(x)})
frequency.head()
```

```python
# monetery
monetery = df.groupby(['account_number']).agg({'lcy_amount': lambda x: sum(x)})
monetery.head()
```
