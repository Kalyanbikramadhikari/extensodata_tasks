```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://election.gov.np/np/page/voter-list-db"

# Fetch the HTML content of the webpage this will be the raw HTML
response = requests.get(url)

# The parser in Beautiful Soup is responsible for taking raw HTML or XML markup and turning it into a parse tree,
#  which is a hierarchical representation of the HTML or XML document. This parse tree allows you to navigate and manipulate the document's elements and attributes easily.
soup = BeautifulSoup(response.text, 'html.parser')

print(soup.prettify())
```
