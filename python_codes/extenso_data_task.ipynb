```python

import sqlalchemy # Python SQL toolkit that allows developers to access and manage SQL databases using Pythonic domain language. You can write a query in the form of a string or chain Python objects for similar queries.
import pandas as pd
```

```python
def dynamic_connection(username, password, hostname,port,database_name, table_name):
    engine = sqlalchemy.create_engine(
        f"mysql+mysqlconnector://{username}:{password}@{hostname}:{port}/{database_name}"
    )

    query = f'SELECT * from {table_name}'

    df = pd.read_sql(query,engine)
    return df
```

```python
# to calculate monthly value lets convert last modified date into datetime field

rw_transaction_data_table['time']= pd.to_datetime(rw_transaction_data_table['time'])
rw_transaction_data_table['last_modified_date']= pd.to_datetime(rw_transaction_data_table['last_modified_date'])




rw_transaction_data_table.dtypes
```

```python
# identify unique product 
result = rw_transaction_data_table.groupby(['module_id', 'product_id', 'payer_account_id']).agg(
    total_amount=('amount', 'sum'),
    total_reward_points=('reward_point', 'sum'),
    total_cash_back=('cash_back_amount', 'sum'),
    total_revenue=('revenue_amount', 'sum'),
    transaction_count=('txn_id', 'count')
).reset_index()
result.head()
```

```python
#making the copy
copy_of_transaction = rw_transaction_data_table
```

```python
# adding new transaction year_month column
copy_of_transaction['year_month']= copy_of_transaction['last_modified_date'].dt.to_period('M')
copy_of_transaction.head()
```

```python
#adding new transaction year-week column
# copy_of_transaction['year_week']= copy_of_transaction['last_modified_date'].dt.to_period('W').astype(str)
# copy_of_transaction.head()

# copy_of_transaction = copy_of_transaction.drop(columns=['year_week'])
copy_of_transaction['year_week']= copy_of_transaction['last_modified_date'].dt.to_period('W').astype(str)
copy_of_transaction.head()
```

```python
monthly_result = copy_of_transaction.groupby(['module_id', 'product_id', 'product_type_id','year_month']).agg(
    total_amount=('amount', 'sum'),
    total_reward_points=('reward_point', 'sum'),
    total_cash_back=('cash_back_amount', 'sum'),
    total_revenue=('revenue_amount', 'sum'),
    transaction_count=('txn_id', 'count')
).reset_index()
monthly_result.head()
```

```python
weekly_result = copy_of_transaction.groupby(['module_id', 'product_id', 'product_type_id','year_week']).agg(
    total_amount=('amount', 'sum'),
    total_reward_points=('reward_point', 'sum'),
    total_cash_back=('cash_back_amount', 'sum'),
    total_revenue=('revenue_amount', 'sum'),
    transaction_count=('txn_id', 'count')
).reset_index()
weekly_result.head()
```

```python
result_inner_join = pd.merge(copy_of_transaction, product_category_map_table , on=['module_id','product_id','product_type_id'], how='inner')
result_inner_join
```

```python
result_inner_join.info()
```

```python
# creating a pivot table
# index is what we want in x-axis or row
# values means what only we want in our result.
# columns means for which column's name we want to set the pivot meaning this column ma bhako diff value will come as row


pivot_table = pd.pivot_table(result_inner_join,values='amount', index=['payer_account_id','year_month'], columns='txn_flow', aggfunc='sum')
pivot_table

#this will now create pivot table with new columns as datas of txn_flow for only amount column 
```

```python
total_pivot_table= pivot_table.reset_index()
total_pivot_table.head()
```

```python
total_pivot_table = total_pivot_table.drop(columns="year_month", axis=0)
total_pivot_table.head()
```

```python
monthly_amount = total_pivot_table.groupby(["payer_account_id"]).mean().reset_index().rename(columns={"InFlow":"Monthly_Inflow_Amount","OutFlow":"Monthly_OutFlow_Amount","Value Chain":"Monthly_Value_Chain"})
monthly_amount.head()
```

```python
total_amount = total_pivot_table.groupby("payer_account_id").sum().reset_index().rename(columns={"InFlow": "Total_inflow_amount","OutFlow":"Total_outflow_amount","Value Chain":"Total_value_chain_amount"})\

total_amount.head()
```

```python
count_pivot_table  = pd.pivot_table(result_inner_join, values='amount', index=['payer_account_id','year_month'], columns='txn_flow', aggfunc='count')
count_pivot_table.head()
```

```python
count_pivot_table = count_pivot_table.reset_index()
count_pivot_table.head()
```

```python
count_pivot_table = count_pivot_table.drop(columns='year_month', axis=1)
count_pivot_table.head()
```

```python
count_pivot_table.dtypes
```

```python
Count_pivot_table_total_amt = count_pivot_table.groupby("payer_account_id").sum().reset_index()\
.rename(columns={"InFlow":"total_Inflow_Count","OutFlow":"total_Outflow_Count","Value Chain": "total_Value_Chain_count"})
Count_pivot_table_total_amt.head()
```

```python
monthly_count = count_pivot_table.groupby(["payer_account_id"]).mean().reset_index().rename(columns={"InFlow":"Monthly_Inflow_Count","OutFlow":"Monthly_Outflow_Count","Value Chain": "Monthly_Value_Chain_count"})\

monthly_count
```

```python
result = result_inner_join.drop_duplicates('payer_account_id')
latest_product = result[['payer_account_id', 'last_modified_date','product_name']].rename(columns={"last_modified_date":"latest_transaction_date","product_name":"latest_used_product"})
latest_product

```

```python
revenue_amount = result_inner_join.groupby(['payer_account_id','year_month']).agg({'revenue_amount':'sum'})
revenue_amount
```

```python
revenue_amount.reset_index()
```

```python
total_revenue = revenue_amount.groupby(['payer_account_id']).agg(total_revenue=('revenue_amount', 'sum'),monthly_revenue=('revenue_amount', 'mean')).reset_index()
total_revenue
```

```python
datas_count = result_inner_join.groupby(["payer_account_id", "product_name"])["product_name"].count().reset_index(name="count")
datas_count_sorted = datas_count.sort_values(by=["payer_account_id", "count"], ascending=[True, False])
most_used_product = datas_count_sorted.groupby("payer_account_id").nth(0).reset_index().rename(columns={"product_name":"most_used_product"})[['payer_account_id', 'most_used_product']]
```

```python
second_most_used_product = datas_count_sorted.groupby("payer_account_id").nth(1).reset_index().rename(columns={"product_name":"second_most_used_product"})\
[['payer_account_id', 'second_most_used_product']]
```

```python
second_most_used_product
```

```python
third_most_used_product = datas_count_sorted.groupby("payer_account_id").nth(1).reset_index().rename(columns={"product_name":"third_most_used_product"})\
[['payer_account_id', 'third_most_used_product']]
```

```python
product_usage = datas_count.groupby("payer_account_id")['product_name'].count().reset_index().rename(columns={"product_name":"product_usage"})
```

```python
product_usage
```

```python
def merged():
    dataframes = [
        reward_point,
        monthly_amount,
        total_amount,
        Count_pivot_table_total_amt,
        monthly_count,
        latest_transaction,
        latest_product,
        total_revenue,
        most_used_product,
        second_most_used_product,
        third_most_used_product,
        product_usage
    ]
    
    merged_df = dataframes[0]
    for dfs in dataframes[1:]:
        merged_df = pd.merge(merged_df, dfs, on='payer_account_id', how='left')
    return merged_df
```
